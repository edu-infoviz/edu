### After some analysis, found that below countries in suicide/life-expectancy are similar, but different than barro lee 
'Congo, D.R.','Cote dIvoire','Dominican Rep.','Hong Kong, China','Republic of Korea','Reunion','Russian Federation','USA'

Renaming these in barrolee data

```python
#Country cleanup in barrolee
barro_lee_data.loc[(barro_lee_data.country == 'Congo, D.R.'),'country']= 'Democratic Republic of Congo'
barro_lee_data.loc[(barro_lee_data.country == 'Cote dIvoire'),'country']= 'Cote d\'Ivoire'
barro_lee_data.loc[(barro_lee_data.country == 'Dominican Rep.'),'country']= 'Dominican Republic'
barro_lee_data.loc[(barro_lee_data.country == 'Hong Kong, China'),'country']= 'Hong Kong'
barro_lee_data.loc[(barro_lee_data.country == 'Republic of Korea'),'country']= 'South Korea'
#assumption based on wikipedia article: https://en.wikipedia.org/wiki/South_Korea
#"Reunion" and "Hong Kong" is not present at all in suicide_death_rate_countries
barro_lee_data.loc[(barro_lee_data.country == 'Russian Federation'),'country']= 'Russia'
barro_lee_data.loc[(barro_lee_data.country == 'USA'),'country']= 'United States'
```

### Renaming a very long column name in suicide data to a simple name

```python
suicide_death_rate_data.rename(columns = {'Deaths - Self-harm - Sex: Both - Age: Age-standardized (Rate) (deaths per 100,000 individuals)':'SuicideRate'},inplace=True)
```

### First merge barro lee with life expectancy and then merge this data with suicide rate
Drop extra column that are generated by the merge like year, country name(Entity) and country code 

```python
merged_data = barro_lee_data.merge(life_expectancy_data,left_on=['country','year'],right_on=['Entity','Year'],how = 'left').reset_index(drop=True)
merged_data.drop(['Year','Entity','Code'],axis=1,inplace=True)
merged_data = merged_data.merge(suicide_death_rate_data,left_on=['country','year'],right_on=['Entity','Year'],how = 'left').reset_index(drop=True)
merged_data.drop(['Year','Entity','Code'],axis=1,inplace=True)
```

### Rate calculations

```python
merged_data.shape
```

```python
merged_data.dropna(subset=['Life expectancy (years)'],inplace=True)
```

```python
merged_data.shape
```

```python
results_country = pd.DataFrame(columns=['country','education_change_rate','life_expectancy_change_rate'])
for country, group_df in merged_data.groupby(['country']):
    group_df.dropna(subset=['tyr'],inplace=True)
    group_df.dropna(subset=['Life expectancy (years)'],inplace=True)
    
    start_year = np.min(group_df['year'])
    end_year = np.max(group_df['year'])
    start_life_expectancy = group_df[group_df['year'] == start_year]['Life expectancy (years)'].iloc[0]
    end_life_expectancy = group_df[group_df['year'] == end_year]['Life expectancy (years)'].iloc[0]
    life_expectancy_change_rate =  (end_life_expectancy - start_life_expectancy)/(end_year - start_year)
    
    start_total_schooling = group_df[group_df['year'] == start_year]['tyr'].iloc[2]
    end_total_schooling = group_df[group_df['year'] == end_year]['tyr'].iloc[2]
    education_change_rate =  (end_total_schooling - start_total_schooling)/(end_year - start_year)
    df = pd.DataFrame({"country":[country], 
                    "education_change_rate":[education_change_rate],  
                    "life_expectancy_change_rate":[life_expectancy_change_rate]}) 
    results_country = results_country.append(df,ignore_index=True)
        
    
```

```python
results_country
```

```python
results_country.to_csv(os.path.join(root_folder,"edu\\data","country_edu_life_expectancy_rates.csv"),header=1,index=False)
```

```python
results_region = pd.DataFrame(columns=['region_code','education_change_rate','life_expectancy_change_rate'])
for region, group_df in merged_data.groupby(['region_code']):
    print(region)
    group_df.dropna(subset=['tyr'],inplace=True)
    group_df.dropna(subset=['Life expectancy (years)'],inplace=True)
    
    start_year = np.min(group_df['year'])
    end_year = np.max(group_df['year'])
    start_life_expectancy = np.mean(group_df[group_df['year'] == start_year]['Life expectancy (years)'])
    end_life_expectancy = np.mean(group_df[group_df['year'] == end_year]['Life expectancy (years)'])
    life_expectancy_change_rate =  (end_life_expectancy - start_life_expectancy)/(end_year - start_year)
    
    start_total_schooling = np.mean(group_df[group_df['year'] == start_year]['tyr'])
    end_total_schooling = np.mean(group_df[group_df['year'] == end_year]['tyr'])
    education_change_rate =  (end_total_schooling - start_total_schooling)/(end_year - start_year)
    df = pd.DataFrame({"region_code":[region], 
                    "education_change_rate":[education_change_rate],  
                    "life_expectancy_change_rate":[life_expectancy_change_rate]}) 
    results_region = results_region.append(df,ignore_index=True)
        
```

```python
results_region.to_csv(os.path.join(root_folder,"edu\\data","region_edu_life_expectancy_rates.csv"),header=1,index=False)
```
